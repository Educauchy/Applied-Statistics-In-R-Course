 ## Fisher's irises

```r
iris <- read.table(file = paste0(path, "iris.txt"), header = F, col.names = c("sepal_l", "sepal_w", "petal_l", "petal_w", "type"), dec = ".", sep = ",")
setosa <- subset(iris, type == "Iris-setosa")
versi <- subset(iris, type == "Iris-versicolor")
virgin <- subset(iris, type == "Iris-virginica")
```

**H0: Test the hypotheses that the correlation between “sepal length” and “sepal width” inside each class of iris is significant. Use Kendall and Spearman correlation coefficients.**

```r
# Setosa
cor.test(setosa$sepal_l, setosa$sepal_w, method = "kendall")
cor.test(setosa$sepal_l, setosa$sepal_w, method = "spearman", exact = F)
plot(setosa$sepal_l, setosa$sepal_w, main = "Setosa. Correlation plot between sepal length and width", xlab = "Length", ylab = "Width")
# Kendll correlation = 0.61 => weak correlation
# Spearman correlation = 0.77 => good correlation
```

```r
# Versicolor
cor.test(versi$sepal_l, versi$sepal_w, method = "kendall")
cor.test(versi$sepal_l, versi$sepal_w, method = "spearman", exact = F)
plot(versi$sepal_l, versi$sepal_w, main = "Versicolor. Correlation plot between sepal length and width", xlab = "Length", ylab = "Width")
# Kendall correlation = 0.398 => no correlation
# Spearman correlation = 0.52 => no correlation
```

```r
# Virginica
cor.test(virgin$sepal_l, virgin$sepal_w, method = "kendall")
cor.test(virgin$sepal_l, virgin$sepal_w, method = "spearman", exact = F)
plot(virgin$sepal_l, virgin$sepal_w, main = "Vigrinica. Correlation plot between sepal length and width", xlab = "Length", ylab = "Width")
# Kendall correlation = 0.31 => no correlation
# Spearman correlation = 0.43 => no correlation
```

```r
# Setosa
cor.test(setosa$petal_l, setosa$petal_w, method = "kendall")
cor.test(setosa$petal_l, setosa$petal_w, method = "spearman", exact = F)
plot(setosa$petal_l, setosa$petal_w, main = "Setosa. Correlation plot between petal length and width", xlab = "Length", ylab = "Width")
# Kendll correlation = 0.18 => no correlation
# Spearman correlation = 0.22 => no correlation
```

## Coins and Bernoulli tests

Suppose in a coin tossing, the chance to get a head or tail is 50%. In a real case, we have 100 coin tossings, and get 48 heads, is our original hypothesis true?

```r
binom.test(x = 48, n = 100, p = 0.5)
# P-value = 0.76 > 0.05 => true probability of success = 0.5, our hypothesis is true
```

Did a fair coin produce 8 heads in 10 flips? By “fair” we mean the coin with equal probabilities of appearance of both sides.

```r
binom.test(x = 8, n = 10, p = 0.5)
# P-value = 0.11 > 0.05 => coin is fair
```
